```python
class Hero:
    def __init__(self,new_name='Без имени'):
        self._name=new_name
        self._x=0
        self._y=0
        self._items=[]
        self._bullets=3
    def location(self):
        print(f"Местонахождение {self._name}: ({self._x},{self._y})")
    def set_name(self, new_name):
        self._name = new_name
    def forward(self):
        if self._y==0:
            if "байк" in self._items:
                self._x+=5
                print(f"{self._name} едет на байке")
            else:
                self._x+=1
                print(f"{self._name} бежит вперед")
        else:
            if "сандали Гермеса" in self._items:
                self._x+=5
                print(f"{self._name} быстро летит")
            else:
                self._x+=1
                print(f"{self._name} летит вперед")
    def back(self):
        if self._y==0:
            if "байк" in self._items:
                self._x-=5
                print(f"{self._name} едет на байке назад")
            else:
                self._x-=1
                print(f"{self._name} бежит назад")
        else:
            if "сандали Гермеса" in self._items:
                self._x-=5
                print(f"{self._name} быстро летит назад")
            else:
                self._x-=1
                print(f"{self._name} летит назад")
    def shoot(self):
        if self._bullets==0:
            print(f"{self._name} потратил все пули")
        else:
            if "пулемет" in self._items:
                print(f"{self._name} стреляет из пулемета")
                self._bullets-=1
                self._x-=1 #отдача
            else:    
                print(f"{self._name} стреляет")
                self._bullets-=1
                self._x-=1
    def collect(self,item):
        if 'умножатель' in self._items:
            if item=='пуля':
                self._bullets+=2
            else:
                self._items.append(item)
        else:
            if item=='пуля':
                self._bullets+=1
            else:
                self._items.append(item)
        print(f"{self._name} собрал плюшку")
class Superhero(Hero):
    def __init__(self, new_name='Супергерой без имени'):
        Hero.__init__(self, new_name=new_name)
    def up(self):
        if 'супергеройский плащ' in self._items:
            print(f"{self._name} взлетает высоко")
            self._y+=5
        else:
            print(f"{self._name} взлетает")
            self._y+=1
    def down(self):
        self._y=0
        print(f"{self._name} приземляется")
```

```python
A=Hero('Иван')
A.forward()
A.shoot()
A.shoot()
A.shoot()
A.shoot()
A.location()
B=Superhero('Вася')
B.up()
B.forward()
B.location()
B.collect('пулемет')
B.shoot()
```
