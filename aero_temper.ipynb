```python
with open('temper.stat.txt','r',encoding='cp1251') as file:
    temp = file.readlines()
def f(str):
    return(float(str[0:-2]))
stat=list(map(f,temp))
c=0
for t in stat:
    if stat.count(t)==1:
        c+=1
print("Максимальная температура: ",max(stat),"\n","Минимальная температура: ",min(stat),"\n","Средняя температура: ",sum(stat)/len(stat),"\n","Количество уникальных температур: ",c)
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
x=list(range(len(stat)))
plt.plot(x,stat)
plt.show()
```
