```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/dm-fedorov/python_basic/master/data/vkontakte_group_01_2016-08-01_2020-03-15.csv")
```

```python
df.head()
```

```python
vi=df[df["Критерий"]=='views']
vs=df[df["Критерий"]=='visitors']
```

```python
vs['год']=vs['Дата']. apply(data)
```

```python
def data(st):
    return st[-4:]
```

```python
vi['год']=vi['Дата']. apply(data)
```

```python
res=vi.groupby("год").sum()
res1=vs.groupby("год").sum()
mean1=vi.groupby("год").mean()
mean2=vs.groupby("год").mean()
```

```python
yy=list(res["Значение"])
yy1=list(res1["Значение"])
yy2=list(mean1["Значение"])
yy3=list(mean2["Значение"])
```
